```python
%matplotlib inline 


import numpy as np #To use functions or constants that are in the NumPy module, type np.function_name. EG: np.pi is pi
import numpy.random as nprd #This imports random number generation functions
import scipy as sp #SciPy has lots of useful numerical methods, so we don't have to write them
import scipy.linalg as spla #This imports the sub-module containing linear algebra methods
import scipy.sparse as sparse
import scipy.sparse.linalg as spsla
import scipy.special as special 
import matplotlib.pyplot as plt #This will let us display images nicely
import cv2
import random
```

```python
### Defining the parameters 
nx=1000
dt=1
a=0
b=20
dx=(b-1)/nx
x=np.linspace(a+dx, b, nx)
diff=1/(6*(dx**2))
perf=(2/3)*(10**(-3))
```

```python
##Initial condition
u=np.zeros(nx)
```

```python
# matrix for the diffusion term 


main_second_der = -2 * np.ones([nx,1])
off_secon_der = np.ones([nx-1,1])
Diff_matrix1 = np.diagflat(main_second_der) + np.diagflat(off_secon_der, -1) + np.diagflat(off_secon_der,1)



main_first_der = np.zeros([nx,1])
off_up_first_der = np.ones([nx-1,1])
off_down_first_der = np.ones([nx-1,1])
for i in range(2,(nx-1)):
    off_up_first_der[i] = off_up_first_der[i]/i

for i  in range( 1,(nx)):
    off_down_first_der[i-1] = - off_down_first_der[i-1]/i

Diff_matrix2= np.diagflat(main_first_der) + np.diagflat(off_down_first_der, -1) + np.diagflat(off_up_first_der,1)

Diff3 = Diff_matrix1 + Diff_matrix2

```

```python
Diff3
```

```python
##Adding in the Neumann condition at the origin

Diff3[1,1] = -6
Diff3[1, 2] = 6

Diff3 = Diff3* diff
Diff3 = sparse.csr_matrix(Diff3)
```

```python
## Adding in the diffusion term
I = sparse.eye(nx);
Perf_matrix = perf * sparse.eye(nx);
```

```python
##Adding in the source term Q
Source = np.sin(x)/x
Source = special.jv(1,x)/x
Source[1] = 1
news =  5* Source**2

news = np.transpose(news)
```

```python
## Timestepping using the Crank-Nicolson method
rapint = u;
tt=np.zeros(60)
trapint=0;
ux=np.zeros(60)
for i in range(1, 61): 
    uold = u
    tt[i-1]=i*dt
    AA = (I - (dt/2)*Diff3 + (dt/2)*Perf_matrix);
    
    if tt[i-1] < 30:
        new_step = (I + (dt/2)*Diff3 - (dt/2)*Perf_matrix).dot(u) + dt*news
    else:
        new_step = (I + (dt/2)*Diff3 - (dt/2)*Perf_matrix).dot(u)
    
     
    u = spsla.spsolve(AA,new_step)
    uu=u
    xx=x;
    xx[nx-1]=b
    uu[nx-1]=0
    #Computing the integral for the damage using trapezoidal rule
    A = 226.8
    E = 6.28*(10^5)
    R = 8.314
    #Function to evaluate
    fold = np.exp(226.8 - E/(R*(uold+310)));
    f    = np.exp(226.8 - E/(R*(u+310)));

    trapint = trapint+0.5*(fold+f)*dt
 
    
    if tt[i-1]< 30:
        plt.plot(xx,uu)
        plt.title('heating up')
        plt.ylabel('Temperature profile')
        plt.xlabel('x')
        plt.show()

    else:
        plt.plot(xx,uu)
        plt.title('cooling down')
        plt.ylabel('Temperature profile')
        plt.xlabel('x')
        plt.show()
    ux[i-1]=uu[0];
    plt.plot(x,trapint)
    plt.title('cooling down')
    plt.ylabel('Damage')
    plt.xlabel('x')
    plt.show()

plt.plot(tt,ux)
plt.ylabel('Temoperature Profile ')
plt.xlabel('Time')
plt.show()
```
